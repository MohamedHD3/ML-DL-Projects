```python
import cv2
import pandas as pd
import os
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
import matplotlib.pyplot as plt
import seaborn as sns
import pickle  # Use pickle for saving the model
```

```python
# Load and preprocess data
data_directory = 'datasets/'
no_tumor_images = os.listdir(os.path.join(data_directory, 'no'))
yes_tumor_images = os.listdir(os.path.join(data_directory, 'yes'))

dataset = []
label = []
```

```python
# Load 'no' tumor images
for filename in no_tumor_images:
    if filename.endswith('.jpg'):
        image_path = os.path.join(data_directory, 'no', filename)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        image = cv2.resize(image, (200, 200))  # Resize to a consistent size
        # dataset.append(image.flatten())  # Use raw pixel values as features
        dataset.append(image)
        label.append(0)  # 0 indicates 'no' tumor
```

```python
# Load 'yes' tumor images
for filename in yes_tumor_images:
    if filename.endswith('.jpg'):
        image_path = os.path.join(data_directory, 'yes', filename)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        image = cv2.resize(image, (200, 200))  # Resize to a consistent size
        # dataset.append(image.flatten())  # Use raw pixel values as features
        dataset.append(image)
        label.append(1)  # 1 indicates 'yes' tumor
```

```python
# Convert lists to numpy arrays
dataset = np.array(dataset)
label = np.array(label)
```

## Split the Data

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dataset, label, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Feature Scaling

```python
print(X_train.max(), X_train.min())
print(X_test.max(), X_test.min())
X_train = X_train/255
X_test = X_test/255
print(X_train.max(), X_train.min())
print(X_test.max(), X_test.min())
```

## Feature Selection: PCA (Principal Component Analysis)

# Models ML

### LogisticRegression

```python
lr_classifier = LogisticRegression(C=0.1)
lr_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_lr_regresor = lr_classifier.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_lr_regresor)
precision = precision_score(y_test, y_pred_lr_regresor)
conf_matrix = confusion_matrix(y_test, y_pred_lr_regresor)
```

```python
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Confusion Matrix:')
print(conf_matrix)
```

### Support Vector Machine (SVM) classifier

```python
# Train a Support Vector Machine (SVM) classifier
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_svc = svm_classifier.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_svc)
precision = precision_score(y_test, y_pred_svc)
conf_matrix = confusion_matrix(y_test, y_pred_svc)
```

```python
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Tumor', 'Tumor'], yticklabels=['No Tumor', 'Tumor'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

## Test new Images

```python
import cv2
import numpy as np
import os
from skimage.feature import hog
from sklearn.preprocessing import StandardScaler
import pickle

# Load the SVM model
with open('lr_model.pkl', 'rb') as model_file:
    lr_classifier = pickle.load(model_file)

# Preprocess the new image without HOG features
def preprocess_image(image_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Resize the image to a consistent size
    image = cv2.resize(image, (64, 64))
    
    # Flatten the image array
    flattened_image = image.flatten()
    
    # Standardize features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(flattened_image.reshape(1, -1))
    
    return scaled_features


folder_path = 'pred'
image_name = 'pred11.jpg'
new_image_path = os.path.join(folder_path, image_name)

# Preprocess the new image
new_image_features = preprocess_image(new_image_path)

# Check the number of features
print("Number of features:", new_image_features.shape[1])

# Make a prediction using the SVM model
prediction = lr_classifier.predict(new_image_features)

# Display the result
if prediction == 1:
    print("The model predicts: Tumor (Class 'yes')")
else:
    print("The model predicts: No Tumor (Class 'no')")
```
